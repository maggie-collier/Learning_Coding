# __Git Notes__

## git status
- shows untracked files, files that need to be committed or added
- paths mentioned are relative to the current directory

## git clone
- clone existing repository into a new directory

## git pull
- integrates with another repository into current branch
- fetches, then merges

## git add
- adds file contents to the index
- prepares for next commit
- `git add .` adds all files in repository

## git commit
- record changes to the repository
- after git add
- create "log" message describing changes

## git log
- show commit logs
- use `git log -n#` to see the last # commits only (ie. `git log -n2` for last 2 logs)

## git push
- updates remote refs using local refs
- after git commit

## git config
- where git settings are stored
- might need to change specific repo settings to fix certain errors

## cd - change directory
- `cd /c/` navigates to c drive
- `cd ~` or `cd` navigates to home directory, users > maggie
- `cd -` navigates to place you were just at
- `cd ..` takes you up a directory, adding / allows you to do this through multiple folders

## ls - list files
- `ls -a` will show all, even hidden files that begin with "."
- `ls -l` shows long format, info shown in file explorer
- `ls -lh` makes size of file 'human readable' instead of bytes

## pwd - current path
- prints the name of the current working directory

## where
- returns location of file
- where Python_Notes.ipynb returns C:\Learning_Coding\Python_Notes.ipynb
- useful to make sure you've updated PATH

## echo - print
- prints what you type to the terminal

## touch - create files
- makes files
- using a space between files can create multiple files

## rm - removes files
- can also do multiple files

## cat - display file contents
- can use -a here too

## vim - text editor from command line
- can use vim in command line to edit documents 
- can make a new file and put something in it
- takes many arguments to do different things
- `vim --file_name` existing or create new files

## >> - append
- `>> --file_name` creates a new file called "file_name"
- arrows take contents of terminal to a file
- used w other commands like echo, `echo hi >> filename`
- `cat file1 >> file2` to move contents from one file to another

## > - overwrite
- overwrite file instead of append

## | - pipe
- pipe is used to combine two or more commands, where the output of the first command is the input for the second, etc
- `echo $PATH | grep "Python"` shows where Python is in PATH

## grep - search for contents
- can search for pattern in each file (ie. ##) and returns every place that has that pattern in the file
- `grep "##" Git_Notes.ipynb` 
- `grep -r "##" .` searches for ## in the current folder and subfolders, r is for recursion

## find - search for files recursively
- `find -iname file` for case insensitive search for name file
- alias `f`

## mkdir - make directory
- makes a new folder/directory

## rmdir - delete directory

## rm - delete file
